```python
import random

# Step 1: Define a list of words
word_list = ['python', 'hangman', 'programming', 'challenge', 'computer', 'random', 'function']

# Step 2: Function to select a random word
def select_random_word(word_list):
    return random.choice(word_list).lower()

# Step 3: Function to display the current state of the word
def display_word_state(word, guessed_letters):
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

# Step 4: Function to check if the word is guessed
def is_word_guessed(word, guessed_letters):
    return all(letter in guessed_letters for letter in word)

# Step 5: Main Hangman game function
def hangman_game():
    # Select a random word
    word_to_guess = select_random_word(word_list)
    guessed_letters = set()
    incorrect_guesses = 0
    max_incorrect_guesses = 6  # Limit on the number of incorrect guesses allowed

    print("Welcome to Hangman!")
    print(display_word_state(word_to_guess, guessed_letters))

    while incorrect_guesses < max_incorrect_guesses:
        guess = input("\nGuess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("Please enter a single alphabetic character.")
            continue

        if guess in guessed_letters:
            print("You have already guessed that letter. Try again.")
            continue

        guessed_letters.add(guess)

        if guess in word_to_guess:
            print("Good guess!")
        else:
            incorrect_guesses += 1
            print(f"Incorrect guess! You have {max_incorrect_guesses - incorrect_guesses} incorrect guesses left.")

        print(display_word_state(word_to_guess, guessed_letters))

        if is_word_guessed(word_to_guess, guessed_letters):
            print("\nCongratulations! You've guessed the word:", word_to_guess)
            break
    else:
        print("\nGame Over! You've run out of guesses. The word was:", word_to_guess)

# Run the Hangman game
hangman_game()
```
